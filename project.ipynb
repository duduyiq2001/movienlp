```python
import tensorflow_decision_forests as tfdf
import tensorflow as tf
from sklearn.metrics import zero_one_loss
new_model = tf.keras.models.load_model('RANDOM_FOREST')
testdata = tf.io.parse_tensor(tf.io.read_file("/content/movienlp/code/mixedinputtest.txt"),out_type=tf.float32)
testlabel = tf.io.parse_tensor(tf.io.read_file("/content/movienlp/code/inputlabeltest.txt"),out_type=tf.float32)
testlabel1 = [1 if a == 1 else 0 for a in testlabel]
print(len(testlabel1))
testlabel1 = tf.convert_to_tensor(testlabel1, dtype=tf.int32)
predict1 = new_model.predict(testdata)
predict1 = [1 if a >=0.5 else 0 for a in predict1]
print(zero_one_loss(testlabel1,predict1))
```
